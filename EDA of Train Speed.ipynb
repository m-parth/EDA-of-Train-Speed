## EXPLORATORY DATA ANALYSIS OF TRAIN SPEED

```python
plt.figure(figsize=(30,10))
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

```python
sns.pairplot(df, hue ='Ratings', palette='bright')
```

```python
sns.pairplot(df, hue ='Avg_Speed', palette='bright')
```

```python
sns.set_style(style='darkgrid')
sns.countplot(data=df,x='Destination')
plt.title("Number of Trains per Station (DESTINATION)")
plt.xlabel("Station")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

```python
sns.set_style(style='darkgrid')
sns.countplot(data=df,x='Origin')
plt.title("Number of Trains per Station (ORIGIN)")
plt.xlabel("Station")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.subplots(figsize =(15,5))
sns.barplot(x='Name', y='Top_Speed',data=df, ci = 0)
plt.title("Top Speed by Train Name")
plt.xlabel("Train Name")
plt.ylabel("Top Speed")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.subplots(figsize =(15,5))
sns.barplot(x='Category', y='Top_Speed',data=df, ci = 0)
plt.title("Top Speed by Train Type")
plt.xlabel("Train Type")
plt.ylabel("Top Speed")
plt.xticks(rotation=90)
plt.show()
```
